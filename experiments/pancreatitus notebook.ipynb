```python
df = pd.read_csv('../data/all_transcrips_ZT.csv')
df.head()
```

```python
measurements = []
for pid in df.PatientID.unique():
    one_p = df[df.PatientID==pid]
    measurements.append(len(one_p))
    print('\n',pid, one_p.Severity_AP.unique(), one_p.ZT.unique(), len(one_p.ZT.unique()))
    print(len(one_p))
```

```python
plt.hist(measurements)
plt.title('number of measurements per patient')
```
